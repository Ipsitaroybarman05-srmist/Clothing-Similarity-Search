```python
from autoscraper import AutoScraper
```

```python

amazon_url='https://www.amazon.in/s?k=clothes&crid=SWMKRSZX5U9L&sprefix=clothes%2Caps%2C500&ref=nb_sb_noss_1'
    
wanted_list=['₹499',' T-Shirts',' 729']
```

```python
scraper=AutoScraper()
result=scraper.build(amazon_url,wanted_list)
print(result)
```

```python
scraper.get_result_similar(amazon_url,grouped=True) #different different products we can see in different different ways
```
